<h2>Oppgave 5</h2>

Lag et program med en funksjon som tar a og b som inn-argumenter og som s√• regner ut arealet og ¬´ytre¬ª omkrets til en figur satt sammen av en rettvinklet trekant og en halvsirkel, se figuren under. Med ¬´ytre¬ª omkrets menes samlet lengde av de sorte strekene. Funksjonen skal returnere arealet og ¬´ytre¬ª omkrets, som s√• skrives til skjerm med passende tekst.

<h4>Programmet beregner areal og ytre omkrets til en halvsirkel og en rettvinklet trekant satt sammen, hvor sirkelens diameter er lik trekantens korteste side</h4>

```python
import numpy as np

a = float(input("Oppgi diameter a [mm] til halvsirkelen:"))
b = float(input("Oppgi lengde p√• siden b [mm] av trekanten:"))

#Beregner areal og omkrets til halvsirkel
areal_HS = (np.pi*(a/2)**2)/2
omkrets_HS = 2*np.pi*a/4

#Beregner areal og omkrets (bidreag til ytre omkrets) til trekanten 
areal_T = a*b/2
omkrets_T = np.sqrt(a**2+b**2)+b

ytre_omkrets = omkrets_HS + omkrets_T
areal = areal_HS + areal_T

print("Figurens totale areal er:", round(areal,2), "mm^2")
print("Figurens ytre omkrets er:", round(ytre_omkrets,2), "mm^2")
```

<h2>Oppgave 6</h2>

Skriv en kode som plotter funksjonen ùëì(ùë•)=‚àíùë•^2‚àí5, for x p√• intervallet [-10,10].

<h4>Programmet plotter funksjonen f(x)=-x^2-5</h4>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
y = -x**2-5

plt.figure(1)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.title("f(x)=-x^2-5") 
```
